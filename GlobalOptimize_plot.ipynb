```python
import os
# import shutil
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
def show1(x:list, file3:str, name="", mod = 0):
    
    font1 = {
        'weight' : 'normal',
        'size'   : 20,
    }
    
    #---------------- acc ----------------#
    file_path = f"./saved_exp_info_0/server_acc/"
    file_path3 = f"./saved_exp_info/server_acc/"
    
    inf = []
    for i in range(len(x)):
        inf.append(np.array(pickle.load(open(file_path + x[i], 'rb')), dtype=object))
    
    plt.figure(figsize=(16, 9)) # 设置整个图像的尺寸（长高分别为6、5个单位）
    plt.title(f"{name}_acc",fontsize=24)
    plt.ylabel('accuracy',fontsize=20)
    plt.xlabel('iteration',fontsize=20)
    
    plt.plot(inf[0],label = "FedAvg")
    plt.plot(inf[1],label = "FedProx")
    plt.plot(inf[2],label = "ShannonTao")
    plt.plot(inf[4],label = "ShannonWeight")
    plt.plot(inf[3],label = "Grad")
    
    plt.plot(np.array(pickle.load(open(file_path3 + file3, 'rb')), dtype=object),label = "pattern_3") # 对比

    plt.legend(loc = "lower right", prop=font1) # 设置信息框    
    
    if mod == 0:
        plt.show()
    else:
        plt.savefig(f"./pic/{name}_acc.png", dpi=300, bbox_inches='tight')
    
    #---------------- loss ----------------#
    file_path = f"./saved_exp_info_0/server_loss/"
    file_path3 = f"./saved_exp_info/server_loss/"
    
    inf = []
    for i in range(len(x)):
        inf.append(np.array(pickle.load(open(file_path + x[i], 'rb')), dtype=object))
    
    plt.figure(figsize=(16, 9)) # 设置整个图像的尺寸（长高分别为6、5个单位）
    plt.title(f"{name}_loss",fontsize=24)
    plt.ylabel('accuracy',fontsize=20)
    plt.xlabel('iteration',fontsize=20)
    
    plt.plot(inf[0],label = "FedAvg")
    plt.plot(inf[1],label = "FedProx")
    plt.plot(inf[2],label = "ShannonTao")
    plt.plot(inf[4],label = "ShannonWeight")
    plt.plot(inf[3],label = "Grad")
    
    plt.plot(np.array(pickle.load(open(file_path3 + file3, 'rb')), dtype=object),label = "pattern_3") #对比
    
    plt.legend(loc = "upper right", prop=font1) # 设置信息框    

    if mod == 0:
        plt.show()
    else:
        plt.savefig(f"./pic/{name}_loss.png", dpi=300, bbox_inches='tight')
```
